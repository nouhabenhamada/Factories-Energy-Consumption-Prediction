# **I. Cement 1**

```python
Cement_1_data = pd.DataFrame()
df = pd.read_csv(os.path.join(factories_dir, "Cement_1" + ".csv"))
Cement_1_data = pd.concat([Cement_1_data, df], ignore_index=True)
Cement_1_data['dateTime'] = pd.to_datetime(Cement_1_data['dateTime'])
Cement_1_data.set_index('dateTime', inplace=True)
Cement_1_data.drop(columns=['hms', 'hms_hours', 'weekDay', 'Date', 'Month', 'weekDayName'], inplace=True)
Cement_1_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
features_to_normalize = ['TotalCons', 'month', 'day', 'hour', 'minute', 'day_of_week', 'FactoryCons']

scaler = MinMaxScaler()
Cement_1_norm = Cement_1_data.copy()
Cement_1_norm[features_to_normalize] = scaler.fit_transform(Cement_1_data[features_to_normalize])

Cement_1_norm.head()
```
